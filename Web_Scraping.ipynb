```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
for i in range(0,541,60):
    url = f'https://sofifa.com/players?offset={i}'
    print(url)
```

```python
NAME = []
AGE = []
OVA = []
POT = []
TEAM_NAME = []
CONTRACT = []
VALUE = []
WAGE = []
TOTAL_STATS = []
```

```python
for i in range(0,541,60):
    url = f'https://sofifa.com/players?offset={i}'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, 'lxml')
arr = []
for i in soup.findAll('td'):
    arr.append(str(i))
for name in arr:
    re.sub('^<td .*</td>|<td.*>|<img.*>|<a.*tooltip="|" href.*</a>|</td>|<.*>','',str(i))
    NAME.append(name)
for age in arr:
    print(re.sub('^<td .*e">|</td>|<td .*</td>|<td.*>|<img.*>|" href.*</a>|</td>|<div.*>|<figure .*>|</figure>|<a .*>|2.*>|</div>','',str(age)))
    AGE.append(age)
for ova in arr:
    print(re.sub('<div.*>|<img.*/>|</figure>|<a.*>|2.*>|</div>|</td>|<td class="col col-ae.*>|<td class="col col-pt.*>|<td .*name">|<td class="col col-vl.*>€|<fig.*>|<td class="col col-wg.*>€|<td class="col col-tt.*>|<td class="col-av.*ar">|<td class="col col-oa.*oa">|<sp.*">|</span>','',str(ova)))
    OVA.append(ova)
for pot in arr:
    print(re.sub('<div.*>|<img.*/>|</figure>|<a.*>|2.*>|</div>|</td>|<td class="col col-ae.*>|<td class="col col-oa.*>|<td .*name">|<td class="col col-vl.*>€|<fig.*>|<td class="col col-wg.*>€|<td class="col col-tt.*>|<td class="col-av.*ar">|<td class="col col-pt.*">|<sp.*">|</span>','',str(pot)))
    OVA.append(pot)
for team in arr:
    print(re.sub('^<td .*>|<fig.*>|<img.*>|<div.*>|</div.*>|<a c.*>|</f.*>|</td>|<a h.*/">|</a>','',str(team)))
    TEAM_NAME.append(team)
for contract in arr:
    print(re.sub('^<td .*>|<fig.*>|<img.*>|<div.*>|</div.*>|<a .*>|</f.*>|</td>|<span.*n>','',str(contract)))
    CONTRACT.append(contract)
for value in arr:
    print(re.sub('<div.*>|<img.*/>|</figure>|<a.*>|2.*>|</div>|</td>|<td class="col col-ae.*>|<td class="col col-oa.*>|<td .*name">|<td class="col col-vl.*>€|<fig.*>|<td class="col col-wg.*>|<td class="col col-tt.*>|<td class="col-av.*ar">|<td class="col col-pt.*>','',str(value)))
    VALUE.append(value)
for wage in arr:
    print(re.sub('<div.*>|<img.*/>|</figure>|<a.*>|2.*>|</div>|</td>|<td class="col col-ae.*>|<td class="col col-oa.*>|<td .*name">|<td class="col col-vl.*>€|<td class="col col-wg.*>€|<fig.*>|<td class="col col-tt.*>|<td class="col-av.*ar">|<td class="col col-pt.*>','',str(wage)))
    WAGE.append(wage)
for total in arr:
    print(re.sub('<div.*>|<img.*/>|</figure>|<a.*>|2.*>|</div>|</td>|<td class="col col-ae.*>|<td class="col col-oa.*>|<td .*name">|<td class="col col-vl.*>€|<td class="col col-wg.*>€|<fig.*>|<td class="col col-tt.*p">|</span>|<td class="col-av.*ar">|<td class="col col-pt.*>','',str(total)))
    TOTAL_STATS.append(total)

```

```python
import pandas as pd

df = pd.DataFrame({"Name": NAME,
                   "Age": AGE,
                   "Overall_Rating": OVA,
                   "Potential": POT,
                   "Team_Name": TEAM_NAME,
                   "Contract_Year": CONTRACT,
                   "Value(M)": VALUE,
                   "Wage(K)": WAGE,
                   "Total_Stats": TOTAL_STATS})

df.head() # Check
df.to_csv('fifa.csv') # Export the data
```
